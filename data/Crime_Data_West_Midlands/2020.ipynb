```python
# Dependencies and Setup
import pandas as pd

# Load the csv files
jan = "2020/2020-01-west-midlands-street.csv"
feb = "2020/2020-02-west-midlands-street.csv"
mar = "2020/2020-03-west-midlands-street.csv"
apr = "2020/2020-04-west-midlands-street.csv"
may = "2020/2020-05-west-midlands-street.csv"
jun = "2020/2020-06-west-midlands-street.csv"
jul = "2020/2020-07-west-midlands-street.csv"
aug = "2020/2020-08-west-midlands-street.csv"
sep = "2020/2020-09-west-midlands-street.csv"
octo = "2020/2020-10-west-midlands-street.csv"
nov = "2020/2020-11-west-midlands-street.csv"
dec = "2020/2020-12-west-midlands-street.csv"

# store into Pandas DataFrames
df1 = pd.read_csv(jan)
df2 = pd.read_csv(feb)
df3 = pd.read_csv(mar)
df4 = pd.read_csv(apr)
df5 = pd.read_csv(may)
df6 = pd.read_csv(jun)
df7 = pd.read_csv(jul)
df8 = pd.read_csv(aug)
df9 = pd.read_csv(sep)
df10 = pd.read_csv(octo)
df11 = pd.read_csv(nov)
df12 = pd.read_csv(dec)

# append the dataframes into one
merge_2020_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12], ignore_index=True)
merge_2020_df
```
