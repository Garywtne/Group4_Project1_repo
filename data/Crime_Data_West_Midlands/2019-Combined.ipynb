```python
# Dependencies and Setup
import pandas as pd

import os

# Load the csv files
jan = "2019/2019-01-west-midlands-street.csv"
feb = "2019/2019-02-west-midlands-street.csv"
mar = "2019/2019-03-west-midlands-street.csv"
apr = "2019/2019-04-west-midlands-street.csv"
may = "2019/2019-05-west-midlands-street.csv"
jun = "2019/2019-06-west-midlands-street.csv"
jul = "2019/2019-07-west-midlands-street.csv"
aug = "2019/2019-08-west-midlands-street.csv"
sep = "2019/2019-09-west-midlands-street.csv"
octo = "2019/2019-10-west-midlands-street.csv"
nov = "2019/2019-11-west-midlands-street.csv"
dec = "2019/2019-12-west-midlands-street.csv"

# store into Pandas DataFrames
df1 = pd.read_csv(jan)
df2 = pd.read_csv(feb)
df3 = pd.read_csv(mar)
df4 = pd.read_csv(apr)
df5 = pd.read_csv(may)
df6 = pd.read_csv(jun)
df7 = pd.read_csv(jul)
df8 = pd.read_csv(aug)
df9 = pd.read_csv(sep)
df10 = pd.read_csv(octo)
df11 = pd.read_csv(nov)
df12 = pd.read_csv(dec)

# append the dataframes into one
merge_2019_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12], ignore_index=True)
merge_2019_df.head()
```

```python
lsoa = pd.read_excel("../LSOA.xlsx", engine='openpyxl')
```

```python
lsoa.head()
```

```python
lsoa = lsoa.rename(columns={"LSOA code (2011)":"LSOA code"})
```

```python
lsoa.head()
```

```python
combined = pd.merge(merge_2019_df, lsoa, on = "LSOA code")
combined.head()
```

```python
del combined['Context']
del combined['Reported by']
del combined['Falls within']
del combined['LSOA name (2011)']
del combined['Overall Index of Multiple Deprivation (IMD) Score']
del combined['Index of Multiple Deprivation (IMD) Rank (where 1 is most deprived)']
del combined['Income Score (rate)']
del combined['Income Rank (where 1 is most deprived)']
del combined['Total population: mid 2015 (excluding prisoners)']
del combined['Dependent Children aged 0-15: mid 2015 (excluding prisoners)']
del combined['Population aged 16-59: mid 2015 (excluding prisoners)']
del combined['Older population aged 60 and over: mid 2015 (excluding prisoners)']
del combined['Working age population 18-59/64: for use with Employment Deprivation Domain (excluding prisoners) ']
```

```python
combined.head()
```

```python
ruc = pd.read_excel("../Rural Urban Classification.xlsx", engine='openpyxl')
ruc.head()
```

```python
ruc = ruc.rename(columns={"LAD11CD":"Local Authority District code (2019)"})
ruc.head()
```

```python
combined1 = pd.merge(combined, ruc, on = "Local Authority District code (2019)")
combined1.head()
```

```python
del combined1['LAD11NM']
del combined1['RUC11CD']
del combined1['Rural population1']
del combined1['Hub towns (rural related) population1']
del combined1['Rural including hub towns (rural & rural related) population1']
combined1.head()
```

```python
la = pd.read_excel("../local authorities.xlsx", engine='openpyxl')
la.head()
```

```python
combined2 = pd.merge(combined1, la, on = "Local Authority District code (2019)")
combined2.head()
```
